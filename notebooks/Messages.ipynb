## Messages per Day and Month

The plot below shows two datasets:
1. Messages per Day  
   *Total summ of messages created at this day.*
2. Avg. Messages per Day each Month  
   *Avergae messages this month per day.*

````python
%matplotlib inline

msgs_per_day = {}
msgs_per_month = {}

# Go through all documents and put the data to
# the data maps created above.
for m in messages_data.clone():
    day = parse_date(str(m.get('timestamp')).split(' ')[0])
    month = date(day.year, day.month, 15)
    
    if day not in msgs_per_day:
        msgs_per_day[day] = 0
    msgs_per_day[day] += 1
    
    if month not in msgs_per_month:
        msgs_per_month[month] = 0
    msgs_per_month[month] += 1

xday = []
yday = []
xmonth = []
ymonth = []

for day, n in msgs_per_day.items():
    xday.append(day)
    yday.append(n)
    
# As you can see here, the first month is excluded
# because its value is 0 at this point.       |
#                                            \/
for month, n in list(msgs_per_month.items())[1:]:
    xmonth.append(month)
    mr = monthrange(month.year, month.month)
    ymonth.append(n / mr[1])

plt.figure(figsize=(14, 8), dpi=80)
plt.plot(xday,yday)
plt.plot(xmonth,ymonth, 'r')
plt.legend(('Messages per Day', 'Avg. Messages per Day each Month'), loc=1)
plt.show()
````

## Message Length over Time

Now, we take a look at the message length on average on each day over time.

````python
msg_lengths = []
msg_dates = []

last_day = 0
c_length = 0
count = 0
for m in messages_data.clone():
    ts = m.get('timestamp')
    c_length += m.get('contentlen')
    count += 1
    if ts.day != last_day:
        last_day = ts.day
        msg_lengths.append(c_length / count)
        msg_dates.append(ts)
        c_length = 0
        count = 0
        
plt.figure(figsize=(14, 8), dpi=80)
plt.plot(msg_dates, msg_lengths)
plt.show()
````

## Attachments

Following chart shows the percentile of messages which have one or more attachments and which have not.

````python
labels = ['has no attachment', 'has attachment']
counts = [0, 0]
msgs_w_attachments_per_day = {}
msgs_w_attachments_per_month = {}

for m in messages_data.clone():
    i = 1 if m.get('attachment') else 0
    counts[i] += 1
    
    ts = m.get('timestamp')
    day = date(ts.year, ts.month, ts.day)
    month = date(ts.year, ts.month, 15)
    
    if day not in msgs_w_attachments_per_day:
        msgs_w_attachments_per_day[day] = [1, i]
    else:
        msgs_w_attachments_per_day[day][0] += 1
        msgs_w_attachments_per_day[day][1] += i
        
    if month not in msgs_w_attachments_per_month:
        msgs_w_attachments_per_month[month] = [1, i]
    else:
        msgs_w_attachments_per_month[month][0] += 1
        msgs_w_attachments_per_month[month][1] += i
    
plt.figure(figsize=(8, 8), dpi=80)
plt.pie(counts, labels=labels, autopct='%1.2f%%')
plt.show()
````

Just for fun, lets plot the percentile of messages which include an attachment per day against the time. So here, you can see the evolution when most parts of messages had an attachment.

````python
xday = list(msgs_w_attachments_per_day.keys())
yday = list(msgs_w_attachments_per_day.values())
yday = [v[1] / v[0] for v in yday]

xmonth = list(msgs_w_attachments_per_month.keys())
ymonth = list(msgs_w_attachments_per_month.values())
ymonth = [v[1] / v[0] for v in ymonth]

fig = plt.figure(figsize=(14, 5), dpi=80)
plt.plot(xday, yday, 'b')
plt.plot(xmonth, ymonth, 'r')
fig.legend(('Proportion of messages containing an attachment', 'Average proportion over a month'), loc=9)
plt.show()
````
